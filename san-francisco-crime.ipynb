```python
train = pd.read_csv('/kaggle/input/sf-crime/train.csv.zip')
test = pd.read_csv('/kaggle/input/sf-crime/test.csv.zip')
train
```

```python
test
```

```python
alldata = pd.concat([train,test])
alldata['Dates'] = pd.to_datetime(alldata['Dates'])
alldata['Year'] = alldata['Dates'].dt.year
alldata['Month'] = alldata['Dates'].dt.month
alldata['Day'] = alldata['Dates'].dt.day
alldata['Hour'] = alldata['Dates'].dt.hour

 
```

```python
alldata2 = alldata.drop(columns = ['Dates', 'Category', 'Id','Descript','Resolution'])
alldata2
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cat_cols = alldata2.columns[alldata2.dtypes == object]
for i in cat_cols:
    alldata2[i] = le.fit_transform(list(alldata2[i]))
```

```python
train2 = alldata2[:len(train)]
test2 = alldata2[len(train):]
train['Category']
train2
```

```python
from sklearn.model_selection import train_test_split
x_train,x_valid,y_train,y_valid = train_test_split(train2,train['Category'],test_size = 0.2, random_state = 100)
y_valid
```

```python
from catboost import CatBoostClassifier
cbc = CatBoostClassifier(verbose = 50, task_type = 'GPU') #
cbc.fit(x_train,y_train,eval_set = (x_valid,y_valid))
result = cbc.predict_proba(test2)
```
